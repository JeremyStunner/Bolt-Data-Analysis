```python
df[['date', 'time']] = df['Order Created Time'].str.split(' ', expand=True)
```

```python
df['hour'] = df['time'].str.split(':',expand=True)[0]
```

```python
df['Unit_VAT'] = df['Unit Price (with VAT)'] - df['Unit Price (Without VAT)']
```

```python
df['Total_VAT'] = df['Total (with VAT)'] - df['Total (without VAT)']
```

```python
df.head()
```

```python
df['Day_of_week'] = pd.to_datetime(df['date']).dt.day_name()
```

```python
df.head(10)
```

```python
deposits = df[df['Title'] == 'Deposit'][['Order ID','Quantity','Restaurant ID','Unit Price (with VAT)','Total (without VAT)','Total_VAT','Bolt Service Fee']]
```

```python
products = df[df['Title'] == 'ProductBottle'][['Order ID','Quantity','Restaurant ID','Unit Price (with VAT)','Total (with VAT)','Total_VAT','Bolt Service Fee']]
```

```python
products.head(10)
```

```python
Unique_orders = products.drop_duplicates(subset='Order ID')
```

```python
# Revenue generated was derived from the Total with VAT column because the Total without VAT seems to be inacurate after inspection
Unique_orders['Revenue'] = Unique_orders['Total (with VAT)']
```

```python
# how many restaurants accepted these orders
df['Restaurant ID'].nunique()
```

```python
# how many orders did each restaurant get
restaurant_orders = df.groupby('Restaurant ID').sum()['Quantity']
```

```python
restaurant_orders
```

```python
restaurant_orders.plot(kind = 'bar')
plt.title('Distribution of Order Quantity by Restaurants')
```

```python
# which restaurants incure the most vat
restaurant_vat = Unique_orders.groupby('Restaurant ID').sum()['Total_VAT']
```

```python
restaurant_vat
```

```python
restaurant_vat.plot(kind = 'bar', color = 'red')
plt.title('Distribution of Restaurants by Total VAT Incured')
```

```python
# which restaurants generated the highest revenue
revenue_generated = Unique_orders.groupby('Restaurant ID').sum()['Revenue']
```

```python
revenue_generated
```

```python
revenue_generated.plot(kind = 'bar', color = 'grey')
plt.title('Distribution of restaurants by revenue generated')
```

```python
# Which restaurants received the most deposit earnings
deposit_earnings = deposits.groupby('Restaurant ID').sum()['Unit Price (with VAT)']
```

```python
deposit_earnings
```

```python
deposit_earnings.plot(kind = 'bar', color = 'Purple')
plt.title('Distribution of restaurants by deposit earnings')
```

```python
# which day of the week has the highest orders
Daily_orders = df.groupby('Day_of_week').sum()['Quantity']
```

```python
Daily_orders
```

```python
Daily_orders.plot(kind = 'bar', color = 'green')
plt.title('Distribution of the Days of the week by order Quantity')
```

```python
# which time of the day has the highest orders
hour_of_day = df.groupby('hour').sum()['Quantity']
```

```python
hour_of_day
```

```python
hour_of_day.plot(kind = 'line', marker = 'o')
plt.title('Distribution of orders by the Hour of the day')
```

```python
# which days of the month had the highest orders
day_of_month = df.groupby('date').sum()['Quantity']
```

```python
day_of_month
```

```python
day_of_month.plot(kind = 'line', color = 'green', marker = 'o')
plt.title('Distribution of Order Quantity by Days of the month')
plt.xticks(rotation=45)
```
